# Aq_pp Conditional Processing Groups

Just like any other programming language, `aq_pp` also supports conditional processing using if, elif, and else and given condition.

## Overview
There are three options that are used for conditional processing, 
* `-if`: Mandantory, open the conditional group and evaluate given expression
* `-elif`: optional, evaluate given expression.
* `-else`: Mandantry, closes the conditional group

Please have the [official documentation](http://auriq.com/documentation/source/reference/manpages/aq_pp.html#conditional-processing-groups) open on your side where you can refer to general structures, syntax and attributes for these options.

### Prerequisites
Readers are expected to be equipped with knowledge of
* aq_pp command
* [input](aq_input.ipynb) and [output](aq_output.ipynb) specification
* other aq_pp options such as [-eval](aq_pp%20-eval.ipynb), [-map](aq_pp%20-map), etc.

### Special Attributes
* `-true`: Evaluates to true
* `-false`: Evaluates to False

* `-skip`: do not output current row
* `-quit`: stop processing entirely
* `-quitafter`: stop processing after the current input record

### if else

If and else are usually used as pairs, and they're essential options for conditional processing group. If clause **always needs to be closed using `-endif` option**.
It's pretty common to be used with `-filt` option, which evaluates the conditional expression given.

In the example below, we will add a new string column `Climate` that stores `warm` for record of April ~ September, and `cold` for months of October ~ March.

```bash
aq_pp -f,+1,eok,qui $file -d $cols -if -filt 'month > 3 && month < 10' -eval S:Climate '"warm"' \
                                    -else -eval Climate '"cold"' -endif
```

<a id='elif'></a>
### elif

Elif is also supported, and can be used just like other programming language.

We will again create new string column `Climate`, in which we'll store each season this time using `-elif`. Season's are
* December ~ Febrary: winter
* March ~ May: Spring
* June ~ September: Summer
* October & November: Fall

```bash
aq_pp -f,+1,eok,qui $file -d $cols -if -filt 'month == 12 || month < 3' -eval S:Climate '"winter"' \
                                    -elif -filt 'month > 3 && month < 6' -eval Climate '"spring"' \
                                    -elif -filt 'month > 6 && month < 10' -eval Climate '"summer"' \
                                    -else -eval Climate '"fall"' -endif
```

<a id='special_attr'></a>
### Special Attributes

Let's take a look at special attributes that you can use. 2 Attributes below sets the boolean value as a constant.

**true/false**<br>

As you can see, when using `-true`, `-if` option will always evaluate to true, resulting outputting "warm" for all of the records.<br>
Feel free to try substituting it with `-false`, and see what happens.

```bash
aq_pp -f,+1,eok,qui $file -d $cols -if -true -eval S:Climate '"warm"' \
                                    -else -eval Climate '"cold"' -endif
```

**Skip**<br>
As the name, this attribute skips the record when meeting the condition.
Let's try skipping all of the records for December ~ March.

```bash
aq_pp -f,+1,eok,qui $file -d $cols -if -filt 'month ==12 || month < 3' -skip -endif
```

**quit/quitafter**<br>

These 2 stops the execution of the command when (`quit`) the given conditions are met, or one row after (`quitafter`) of that.

Here we will use row numbers to quit outputting the record after 20th records.

```bash
aq_pp -f,+1,eok,qui $file -d $cols -eval I:Index '$rowNum' -if -filt '$rowNum > 20'  -quit -endif
```

quitafter will stop executing right after the 20th record.

```bash
aq_pp -f,+1,eok,qui $file -d $cols -eval I:Index '$rowNum' -if -filt '$rowNum > 20'  -quitafter -endif
```
