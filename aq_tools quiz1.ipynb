### Q1. csv files
using aq_pp command, import the titanic data with proper column spec and display on stdout. 

```bash
pwd
titanic="data/titanic.csv"
cols=$(loginf -f,auto $titanic -o_pp_col -)
aq_pp -f,+1 $titanic -d $cols
```

### Q5. Data filteration
This one's bit tricky. Without usiing Sex column, extract the record of men who survived. Observe how the name column is formatted. 


```bash
aq_pp -f,+1 $titanic -d $cols -eval s:pronoun name \
-map,pcre pronoun "(^M(rs?|is{2}))" '%%1%%' \
-filt '(pronoun=="Mr")&&(Survived==1)' -c survived Pronoun Name Sex  
```
